#Exercício 9

```r
# Resposta 9
boxplot(clarity ~ cut , data = diamantes, xlab = "Corte", ylab = "Profundidade", main = "Exercicio 9", col="orange")
```

```r
boxplot(price ~ color, data = diamantes, xlab = "Cores", ylab = "Preco", main = "Exercicio 9", col="orange")
```

```r
boxplot(price ~ clarity, data = diamantes, xlab = "Claridade", ylab = "Preco", main = "Exercicio 9", col="orange")
```

```r
par(mfrow=c(3,1), mar=c(4,4,2,1))
hist(subset(diamantes, cut=="Very Good")$depth , col="purple", main="", xlab="Very Good", ylab="Preco", mar=c(4,4,2,1))
hist(subset(diamantes, cut=="Ideal")$depth , col="purple", main="", xlab="Ideal", ylab="Preco", mar=c(4,4,2,1))
hist(subset(diamantes, cut=="Premium")$depth , col="purple", main="", xlab="Premium", ylab="Preco", mar=c(4,4,2,1))
```
